```python
real_work_by_user =df.fillna('Software Development').groupby(['User','Group',  'productive'])['Duration (decimal)'].sum().unstack().fillna(0)
real_work_by_user
```

```python
real_work_by_user.columns = ['not real work', 'real work']
```

```python
real_work_by_user['total'] = real_work_by_user['not real work'] + real_work_by_user['real work']
real_work_by_user['real work (%)'] = (real_work_by_user['real work'] / real_work_by_user['total']) * 100
```
